```python
#3번
import pandas as pd
import heapq
import os

def load_and_print_heap():
    
    file_name = 'birthday.csv'
    
    if not os.path.exists(file_name):
        print(f"오류: '{file_name}' 파일을 찾을 수 없습니다.")
        print("VSC에서 '폴더 열기'로 바탕화면을 선택했는지 확인해주세요.")
        return

    df = pd.read_csv(file_name)
    df.columns = [c.strip() for c in df.columns] # 제목 공백 제거

    df = df.dropna(subset=['생년(예. 1985)', '생월(예. 12)', '생일(예. 5)'])

    heap = []
    for _, row in df.iterrows():
        name = row.iloc[0] 
       
        birth_val = int(row['생년(예. 1985)']) * 10000 + \
                    int(row['생월(예. 12)']) * 100 + \
                    int(row['생일(예. 5)'])
        
        
        heapq.heappush(heap, (-birth_val, name, row['생년(예. 1985)'], row['생월(예. 12)'], row['생일(예. 5)']))

   
    print("### 과제 3: 생일이 느린 순서 10명 ###")
    for i in range(1, 11):
        if not heap: break
        val, name, y, m, d = heapq.heappop(heap)
        print(f"{i}위: {name} ({int(y)}년 {int(m)}월 {int(d)}일)")
load_and_print_heap()
```

```python
#4번
import pandas as pd

class Node:
    def __init__(self, item):
        self.item = item
        self.next = self.prev = None

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = Node("dummy")
        self.head.next = self.head.prev = self.head

    def append(self, item):
        new_node = Node(item)
        last = self.head.prev
        new_node.next, new_node.prev = self.head, last
        last.next = self.head.prev = new_node
try:
    df = pd.read_csv('birthday.csv')
    df.columns = [c.strip() for c in df.columns]
    df = df.dropna(subset=['생년(예. 1985)', '생월(예. 12)', '생일(예. 5)'])
    cdll = CircularDoublyLinkedList()
    for _, row in df.iterrows():
        cdll.append(row.tolist())
    my_team = [
        "박다인", "안예원", "정서하", "김우현", "양시언", 
        "유가현", "이하연", "장예은", "전은빈", "정세희", 
        "이예은", "황다원"
    ]
    
    print("### 우리 조원 정보 출력 ###")
    curr = cdll.head.next
    while curr != cdll.head:
        name = curr.item[0]
        if name in my_team:
         
            print(f"이름: {name} | 생일: {int(curr.item[1])}년 {int(curr.item[2])}월 {int(curr.item[3])}일")
        curr = curr.next

except FileNotFoundError:
    print("birthday.csv 파일을 찾을 수 없습니다.")
```
