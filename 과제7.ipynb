B-트리의 한 노드가 가질 수 있는 최대 키의 수를 $m$이라 할 때, 노드 하나를 저장하는 데 필요한 크기는 **(키의 크기 $\times$ 키의 개수) + (포인터의 크기 $\times$ 포인터의 개수)**이다. 이 합계가 디스크 한 블록의 크기인 8,192바이트보다 작거나 같아야 하므로 다음과 같은 식이 성립한다.1) 설계 식$$(20 \times m) + (4 \times (m + 1)) \le 8,192$$2) 계산 과정$20m + 4m + 4 \le 8,192$$24m + 4 \le 8,192$$24m \le 8,188$$m \le 341.16...$3) 결론따라서 B-트리의 한 노드는 최대 341개의 키를 가질 수 있도록 설계한다.




16. 각 노드가 최대 k개의 키를 가질떄 전체 키의 수는 N이라 하면 포인터의 개수는 N+1이다. 

17. 발생하지 않는다. 언더플로우는 키 개수가 [m/2]-1미만일 때 발생하는데 두 형제 노드를 합치고 부모의 키 1개를 가져와도 결과 노드의 키 개수는 [m/2]-1 + [m/2]-1 정도 이므로 최대용량인 m-1을 초과하지 않는다. 

18. 1조 개의 데이터를 저장할 때, 노드 하나가 1,000개의 키를 담을 수 있는 B-트리의 높이는 다음과 같이 결정된다.

1) 가장 낮은 높이 (최소 높이)

모든 노드가 빈틈없이 꽉 채워져서 한 번에 최대한 많은 자식으로 뻗어 나가는 경우이다. 이때는 약 5층 정도면 1조 개의 데이터를 모두 담을 수 있다.

2) 가장 높은 높이 (최대 높이)

모든 노드가 최소한의 데이터(절반)만 채워져서 트리가 위아래로 길게 늘어지는 경우이다. 이때는 꽉 채웠을 때보다 층수가 늘어나 약 6층 정도의 높이가 된다.

3) 결론

따라서 이 B-트리의 높이는 데이터가 채워진 정도에 따라 5에서 6 사이의 값을 가진다.
