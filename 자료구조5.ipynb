```python
#3
import csv
class MaxHeap:
    def __init__(self):
        self.__items = []
    def insert(self, item):
        self.__items.append(item)
        self.__percolateUp(len(self.__items) - 1)
    def __percolateUp(self, i):
        parent = (i - 1) // 2
        if i > 0 and self.__items[i][1] > self.__items[parent][1]:
            self.__items[i], self.__items[parent] = self.__items[parent], self.__items[i]
            self.__percolateUp(parent)
    def deleteMax(self):
        if not self.__items: return None
        max_val = self.__items[0]
        self.__items[0] = self.__items[-1]
        self.__items.pop()
        self.__percolateDown(0)
        return max_val
    def __percolateDown(self, i):
        child = 2 * i + 1
        right = 2 * i + 2
        if child < len(self.__items):
            if right < len(self.__items) and self.__items[child][1] < self.__items[right][1]:
                child = right
            if self.__items[i][1] < self.__items[child][1]:
                self.__items[i], self.__items[child] = self.__items[child], self.__items[i]
                self.__percolateDown(child)
heap = MaxHeap()
try:
    with open('birthday.csv', mode='r', encoding='utf-8') as f:
        reader = csv.reader(f)
        next(reader)
        for row in reader:
            if len(row) >= 4:
                name = row[0].strip()
                year = row[1].strip()
                month = row[2].strip().zfill(2) 
                day = row[3].strip().zfill(2)   
                if year and month and day:
                    bday_num = int(year + month + day) 
                    heap.insert((name, bday_num))
except FileNotFoundError:
    print("birthday.csv 파일을 찾을 수 없습니다.")
print("##생일이 느린 순서 10명 ##")
for i in range(1, 11):
    res = heap.deleteMax()
    if res:
        date_str = str(res[1])
        print(f"{i}. {res[0]} {date_str[:4]}-{date_str[4:6]}-{date_str[6:]}")
    else:
        break
```

```python
#4
import csv 
class Node:
    def __init__(self, item):
        self.item = item
        self.prev = self.next = self
class CircularDoublyLinkedList:
    def __init__(self):
        self.head = Node(None)
    def append(self, item):
        new_node = Node(item)
        last = self.head.prev
        new_node.next = self.head
        self.head.prev = new_node
        new_node.prev = last
        last.next = new_node
my_team = ["황다원", "박다인", "안예원", "정서하", "김우현", "양시언", 
           "유가현", "이하연", "장예은", "전은빈", "정세희", "이예은"]
cdll = CircularDoublyLinkedList()
try:
    with open('birthday.csv', mode='r', encoding='utf-8') as f:
        reader = csv.reader(f)
        next(reader)  
        for row in reader:
            if len(row) >= 4:
                name = row[0].strip()
                if row[1].strip() and row[2].strip() and row[3].strip():
                    bday = f"{row[1]}-{row[2].zfill(2)}-{row[3].zfill(2)}"
                else:
                    bday = None  
                cdll.append({"name": name, "birth": bday})
except FileNotFoundError:
    print("birthday.csv 파일을 찾을 수 없습니다. 파일명을 확인해주세요.")
except Exception as e:
    print(f"오류 발생: {e}")
print("## 조원 생일 ##")
curr = cdll.head.next
found = False
while curr != cdll.head:
    if curr.item["name"] in my_team:
        bday_display = curr.item['birth'] if curr.item['birth'] else "생년월일 정보 없음"
        print(f"이름: {curr.item['name']}, 생년월일: {bday_display}")
        found = True
    curr = curr.next
if not found:
    print("조원 명단에 있는 이름이 데이터에 없습니다.")
```
